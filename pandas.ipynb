```python
##sample
import pandas as pd
import numpy as np
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman',np.nan],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip',np.nan],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT,pd.NaT]})
df
##NaN - not a number , NaT - not a time
```

```python
##drop rows with all null values
df.dropna(how='all')
```

```python
##drop rows if any one valuie is null
df.dropna()
```

```python
##drop rows if 2 or more null value is there
df.dropna(thresh=2)
```

```python
##drop rows if null is present in  any of specified columns
df.dropna(subset=['name','toy'])
```

##FILLNA

```python
df2=df.copy()
df2['toy']=df2['toy'].fillna('No info')
df2
```

```python
df2=df.copy()
df2.ffill()
##fills forward
```

```python
df2=df.copy()
df2.bfill()
##fills backward
```

```python
df2=df.copy()
df2.fillna({'name':'name not found','toy':'not interested'})
##fills each column by custom
```

```python
df2['A']=df2['A'].fillna(df2['A'].mean())
##fill missing values with mean,median,mode
```

##Concat

```python
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
print(df1)
print(df2)
```

```python
##vertical
df3=pd.concat([df1,df2],axis=0)
df3
```

```python
##horizontal
df3=pd.concat([df1,df2],axis=1)
df3
```

##PIVOT

Basic

```python
import pandas as pd

# Sample DataFrame
data = {
    'date': ['2025-01-01', '2025-01-02', '2025-01-01', '2025-01-02'],
    'city': ['New York', 'New York', 'Los Angeles', 'Los Angeles'],
    'temperature': [32, 30, 75, 70]
}
df = pd.DataFrame(data)
print(df)
# Pivot the DataFrame
pivot_df = df.pivot(index='date', columns='city', values='temperature')
pivot_df
```

pivot table with aggregation

```python
# Sample DataFrame
data = {
    'date': ['2025-01-01', '2025-01-01', '2025-01-02', '2025-01-02'],
    'city': ['New York', 'New York', 'Los Angeles', 'Los Angeles'],
    'temperature': [32, 30, 75, 70]
}
df = pd.DataFrame(data)
print(df)

pivot_table_df = pd.pivot_table(df, values='temperature', index='date', columns='city', aggfunc='mean', margins=True)
pivot_table_df
```

##pivot with multiple aggregation

```python
pivot_table_df2 = pd.pivot_table(df, values='temperature', index='date', columns='city', aggfunc=['mean','sum','max'])
pivot_table_df2
```

###melt

```python
import pandas as pd

# Sample DataFrame
data = {
    'id': [1, 2, 3],
    'A': [10, 20, 30],
    'B': [15, 25, 35]
}
df = pd.DataFrame(data)
print(df)
# Melt the DataFrame
df_melted = pd.melt(df, id_vars=['id'], value_vars=['A', 'B'], var_name='variable', value_name='value')

print(df_melted)
```
